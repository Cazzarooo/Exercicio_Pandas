```python
import pandas as pd
import re
```

```python
df = pd.read_csv('/content/house.csv')
```

```python
df
```

```python
# Filtrando com condições
filter1 = df.loc[(df['Rooms'] == 3) & (df['Type'] == 'h') & (df['Price'] <= 500000)]
```

```python
filter1
```

```python
# Filtrando com Regular Expression
filter2 = df.loc[df['Address'].str.contains('Turner st|Turner Rd', flags=re.I)]
```

```python
filter2
```

```python
# Filtrando com Regx e Conditions
filter3 = df.loc[df['Address'].str.contains('^59', flags=re.I) | (df['Price'] <= 500000)]
```

```python
filter3
```

```python
# Modificando dados do dataframe
df.loc[df['SellerG'] == 'Ronald', ['Method', 'Type']] = 'Pending'
```
